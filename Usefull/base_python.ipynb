```python
# "вытягивание" в линию и смена размерности (явная и неявная)
d2.ravel(), d2.reshape(1, 6), d2.reshape(1, -1)
```

```python
# смена размерности ~ "нарезание" на 3
d2.reshape(3, -1)
```

```python
# добавление размерности
d2.reshape(2, 3, 1).shape, d2[:, :, None].shape, d2[..., None].shape
```

```python
# транспонирование — обратный порядок размерностей
d2.T  # ≈ np.transpose(x, (n-1, n-2, ..., 0))
```

### Особые матрицы

```python
# матрица 2х3 из единиц
np.ones((2, 3))
```

```python
# матрица 2х3 из нулей
np.zeros((2, 3))
```

```python
# единичная матрица 3х3, потому что "eye" звучит как "I"
np.eye(3)
```

```python
# диагональная матрица из вектора b
np.diag(b)
```

```python
# матрица 2х3 из случайных чисел равномерного распределения [0, 1)
np.random.rand(2, 3)
```

```python
# выделение неиспользуемой памяти для матрицы 3х2
np.empty((3, 2))
```

```python
# выделение памяти по подобию
empty = np.empty_like(d2, dtype=float)
empty
```

### Список -> Массив -> Тензор

```python
import torch
```

```python
# конвертация в тензор
# a3 = torch.tensor(a)
b3 = torch.tensor(b)
c3 = torch.tensor(c)
d3 = torch.tensor(d)
b3, c3, d3
```

```python
# многомерные индексы
d3[0, -1]
```

```python
# многомерный срезы
d3[:, :2]
```

```python
# меньше циклов - лучше
#b = [i for i in range(8, 2, -1)]
b3 = torch.arange(8, 2, -1)
b3
```

### Операции (as Numpy)

```python
# бродкастинг и неявная конвертация
# d2 + [0, 1, -3]
d3 + np.array([0, 1, -3])
```

```python
# статистики как метод
d3.min(dim=1)  # np.min(d2)
```

```python
# т.е. сразу и np.min, и np.argmin
d3.min(dim=1).values
```

```python
# мат. операции как метод
d3.log()
d3.sin()
d3.exp()
```

```python
# операции inplace, т.е. используется та же память (исходник затирается)
print(d3)
print(d3.add_(1))
print(d3.add_(3))
d3 = torch.tensor(d)
```

### Представления (as Numpy)

```python
# смена типов как astype и смена устройства
d3.to(float), d3.to(torch.int8), d3.to('cpu')
```

```python
# "вытягивание" в линию
d3.ravel()
```

```python
# смена размерности (может создать новый тензор)
d3.reshape(1, -1)
```

```python
# добавление размерности
d3.reshape(2, 3, 1).shape, d3[..., None].shape, d3.unsqueeze(2).shape
```

```python
# представления (гарантирует переиспользование памяти)
d3.view(1, -1), d3.view(torch.int32)
```

```python
# перестановка всех размерностей нескольких
d3.permute(1, 0)
```

```python
# перестановка двух размерностей
d3.transpose(1, 0)
```

```python
# тензорное транспонирование (обратный порядок размерностей)
d3.T  # ≈ x.permut(n-1, n-2, ..., 0)
```

```python
# простое транспонирование (размерность ≤ 2)
d3.t()  # ≈ x.transpose(1, 0)
```
